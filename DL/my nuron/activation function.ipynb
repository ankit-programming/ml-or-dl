activation function ka kam loss , nuron ko triger kerna etc hota hai

ther are diffrernt type of activation function



```python
#nuron
import numpy as np
X = np.random.randn(3,4)

class tensor:
    def __init__(self,n_input,n_nurons):
        self.weight = np.random.randn(n_input,n_nurons)
        self.biases = np.zeros((1,n_nurons))
    def nuron(self,X_input):
        self.output = np.dot(X_input,self.weight) + self.biases

layer1 = tensor(4,2)
#print("weight",layer1.weight)
#print("biases",layer1.biases)
layer1.nuron(X)
#print(">>>",layer1.output)
```
